```python
#import modules

import pandas as pd 
import os 
# %matplotlib inline
from PIL import Image
import numpy as np 
from os import walk

def change(string):
    result = ''
    for i in range(len(string)):
        if string[i] == "\\":
            result += '/'
        else:
            result += string[i]
    return result[2:] if result[0] == 'C' else result


def add_path(number_list,before): #absolute path
    for i in range(len(number_list)):
        number_list[i] = before+'/'+str(number_list[i])
    return number_list
```

```python
#functions to remove outliers. r_outliers gives a list with all outliers and filtering gives a list of the rest of the images
#the desired_ratio basically is the percentage of the 256x256 grid you want the image to cover
#for example, a desired_ratio of 0.4 means that every image that does not cover 40 procent of a square grid will be eliminated
#besides removing too narrow images, too small images will also be removed
#max_size is the maximum length of one of the two sides of an image
#any image with a height OR width less than max_size will be removed

#walk function, note for myself:
#for(Name of the directory, List of subdirectories underneath, List of files underneath) in walk(directory)

def r_outliers(folder_dir, max_size, desired_ratio):
    outliers= []
    filtered_data = []
    for (dirpath, dirnames, filenames) in walk(folder_dir): # folder_dir = database in our case
        dirnames = add_path(dirnames,folder_dir) 
        for folders_insect in dirnames: #for each folder in the list of all subdirectories of database
            folder_name = folders_insect.replace(folder_dir+'/', '') #to get the folder name aka name of insect later
            for (dirpath1, dirnames1, filenames1) in walk(folders_insect): # inside one of the folders
                for pictures in filenames1: #loop through pictures of an insect(=folder_insect)  
                    pic_path = dirpath1 + '/' + pictures  #path of each individual picture
                    sig_pic_path = pic_path.replace(folder_dir, "")  #picture path without redundant main(/database) directory
                    im = Image.open(pic_path) #open/acces image
                    im_size = im.size  #get dimensions (height and width) of image
                    ratio = min(im_size)/max(im_size)  #calculate ratio of image
                    if (im_size[0]<max_size or im_size[1]<max_size) or ratio<desired_ratio: #see first 4 lines
                        outlier = [] #list refreshes each loop to get [(im_size), 'filepath', 'folder', 'img.jpg']  for all outliers
                        outlier.append(im_size)
                        outlier.append(sig_pic_path) 
                        outlier.append(folder_name)
                        outlier.append(pictures) #for new folder creating purposes, see codeblock below (pictures is row[3] in the code below)
                        outliers.append(outlier) #append [(im_size), 'filepath', 'folder', 'img.jpg']  to the outliers list
    return outliers


def filtering(folder_dir, max_size, desired_ratio):
    outliers= []
    filtered_data = []
    for (dirpath, dirnames, filenames) in walk(folder_dir): 
        dirnames = add_path(dirnames,folder_dir) 
        for folders_insect in dirnames: 
            folder_name = folders_insect.replace(folder_dir+'/', '') 
            for (dirpath1, dirnames1, filenames1) in walk(folders_insect): 
                for pictures in filenames1: 
                    pic_path = dirpath1 + '/' + pictures 
                    sig_pic_path = pic_path.replace(folder_dir, "")  
                    im = Image.open(pic_path) 
                    im_size = im.size  
                    ratio = min(im_size)/max(im_size)  
                    if (im_size[0]<max_size or im_size[1]<max_size) or ratio<desired_ratio: 
                        outlier = [] 
                        outlier.append(im_size)
                        outlier.append(sig_pic_path) 
                        outlier.append(folder_name)
                        outlier.append(pictures) 
                        outliers.append(outlier) 
                    else:
                        filtered = [] 
                        filtered.append(im_size)
                        filtered.append(sig_pic_path)
                        filtered.append(folder_name)
                        filtered.append(pictures) 
                        filtered_data.append(filtered) 
    return filtered_data        #this function has 7 lines more which contains an else statement for the rest of the images
```

```python
path = r'C:\Users\elanu\Desktop\data\Insect image dataset (kaggle)\database'
path = './database'
folder_dir = change(path) #createing working dir variable for for loop
```

```python
#outlier = (height OR width)<120 OR ratio<0.33
#120 chosen through trial and error
#ratio = min(h_or_w)/max(h_or_w)   = longestSide/shorterSide
#ratio<0.33 : So the image should at least cover 1/3 of the 256x256 grid aka LongestSideImage < 2*ShorterSide 

outliers = r_outliers(folder_dir, 120, 0.33)
# print(outliers[:5])
# print()
# print(len(outliers), 'outliers')
# print()
# print()
filtered_data = filtering(folder_dir, 120, 0.33)
# print(filtered_data[:5])
# print()
# print(len(filtered_data), 'filtered images')  
```

```python
#putting outliers in a new folder 'outliers'

path_outliers = change(r'C:\Users\elanu\Desktop\data\Insect image dataset (kaggle)\outliers') #created map outliers
path = './outliers'
outliers_images = [] #append all images from outliers list to this list

for row in outliers: #looping through the list outliers
    im_pth = folder_dir + row[1] #paths of all images in outliers list
    im = Image.open(im_pth)  #open/acces image
    outliers_images.append(im) #add the opened image outliers_images to create a list with only the photos
    newpath = path_outliers + '/' + row[2] #define a path with coresponding folder but in another directory named outliers instead of database
    if not os.path.exists(newpath): #check whether there is a folder in the outliers directory named row[2]
        os.makedirs(newpath) #create the newpath if it does not exist, this way we create the same folder names in this new directory
    im.save('{}{}{}'.format(newpath,'/', row[3])) #save the image in this new path with the same name (row[3]) as original image
    
```

```python
#creating resized images and adding to the new folders
desired_size = 256
filtered_and_resized_images = []
path_filtered_and_resized_images = change(r'C:\Users\elanu\Desktop\data\Insect image dataset (kaggle)\filtered and resized images')
path_filtered_and_resized_images = './filtered'

for row in filtered_data:
    old_size = row[0]
    im_pth = folder_dir + row[1]
    im = Image.open(im_pth)
    ratio = float(desired_size)/max(old_size)
    new_size = tuple([int(x*ratio) for x in old_size]) 
    im = im.resize(new_size, Image.ANTIALIAS) 
    new_im = Image.new("RGB", (desired_size, desired_size))
    new_im.paste(im, ((desired_size-new_size[0])//2, (desired_size-new_size[1])//2))   
    filtered_and_resized_images.append(new_im)
    newpath = path_filtered_and_resized_images + '/' + row[2]
    if not os.path.exists(newpath):
        os.makedirs(newpath)
    new_im.save('{}{}{}'.format(newpath,'/', row[3])) #only difference with the outliers version of this code is that
    #in the code for the outliers I just saved the original image in the folder. Here, I am saving the 
    #new_im (aka the resized image) to the corresponding folder with the corresponding filename in the filtered db directory
```

```python
#source: https://stackoverflow.com/questions/49649215/pandas-image-to-dataframe/49650363#49650363
#create {table:[height, width, r, g, b]} for all resized images and append them to one list values
#loading these codes for the whole database did not work, I waited hours, my browsers are crashing. 
#I keep getting issues with setup up the GPU, I messaged the professor and am waiting on his response.
#for now I'll do this for the first 100 files
rsz_filt_pth = change(r'C:\Users\elanu\Desktop\data\Insect image dataset (kaggle)\filtered and resized images')
# rsz_filt_pth = './filtered'
values = []
c = 0


for row in filtered_data:
#     if c == 3:
#         break
    colourImg = Image.open(rsz_filt_pth + row[1])
    colourPixels = colourImg.convert("RGB")
    colourArray = np.array(colourPixels.getdata()).reshape(colourImg.size + (3,))
    indicesArray = np.moveaxis(np.indices(colourImg.size), 0, 2) #changes (2, 256, 256)(=colourImg.size) to (256, 256, 2)
#     print(indicesArray) #indicesArray.shape = (256, 256, 2)
    allArray = np.dstack((indicesArray, colourArray)).reshape((-1, 5))
    value = [] #[array, 'folder/insectname' , '/folder/img.jpg']
    value.append(allArray)
    value.append(row[2])
    value.append(row[1])
    values.append(value)
#     c += 1
    
```

```python
#Here we will insert all allArrays from values into one DataFrame 
#and we will add two columns to this final DataFrame:'Insect species' and 'Filepath'
#again: this is only for the first 100 files/images

df_list = [] #create an empty list
comb_df = pd.DataFrame(df_list) #created empty list to convert it to an empty DataFrame

for value in values:
    df = pd.DataFrame(value[0], columns=["height", "width", "red","green","blue"])  #value[0] = allArray
    df.loc[df.index[:], 'Insect species'] = value[1]    #value[1] = folder/insect_name
    df.loc[df.index[:], 'Filepath'] = value[2]   #value[2] = filepath
    comb_df = (pd.concat([comb_df, df], axis=0)) #combine current df with df from loop

# print(comb_df[:360000])
# print()
# print(len(comb_df)) #should be 4,085,383,168 = 256*256*62338 (heigh*width*amount_of_images)
```
